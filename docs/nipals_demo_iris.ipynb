```python
from nipals import nipals
```

```python
import seaborn as sns

iris = sns.load_dataset('iris')
iris.head()
```

```python
iris = iris.set_index('species')
```

```python
# Initialize class variables
iris['setosa'] = 0
iris['versicolor'] = 0
iris['virginica'] = 0
# Set class variables based on index species
iris.loc['setosa', 'setosa'] = 1
iris.loc['versicolor', 'versicolor'] = 1
iris.loc['virginica', 'virginica'] = 1
iris = iris.set_index(['setosa', 'versicolor', 'virginica'], append=True)
```

```python
iris.head()
```

```python
nip = nipals.Nipals(iris)
nip.fit(ncomp=2, cv=True)
_ = nip.plot(classlevels=['setosa', 'versicolor', 'virginica'], labels='species')
```
